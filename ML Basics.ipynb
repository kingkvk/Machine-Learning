```python
#import the required libraries if any of the libraries missing we can use
#!pip install "Required Library"
#=========================================================================
#Tensorflow library for the machine learning
import tensorflow as tf
#Numerical python Library to handle with arrays
import numpy as np
#Matplot lib to plot the Graphs
import matplotlib.pyplot as plt
#Keras Library is Defining the Model
from tensorflow import keras
```

```python
#creating the data for the machine learning problem
x=np.arange(-5,5)
print(x)
```

```python
#Creating the data as the target output
y=(2*x)+1
print(y)
```

```python
#Defining the keras model with input shape as one
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
#Compiling the model
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
#training the model for 500 epochs
model.fit(x, y, epochs=500)
```

```python
#listing the model layers and the parameters
model.summary()
```

```python
#checking with the trained model
#y=(10*2)+1
print(model.predict([100]))
```

```python
plt.plot(x,y)
plt.xlabel('x data')
plt.ylabel('y data')
plt.show()
```

```python
#saving the trained model
model.save('2*x+1.h5')
```
